```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('creditcard.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
dataset['Class'].value_counts()
```

Data is highly unbalanced, so accuracy isn't a good metric. We can use Reciever Operating Curve AUC (ROC AUC)

```python
dataset_copy = dataset.drop(['Class'],axis = 1)
dataset_copy = np.asarray(dataset_copy)
```

```python
#Plotting KMeans to Visualize the Anomalies
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,init = 'k-means++')
    kmeans.fit(dataset_copy)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#From the Elbow method, we got 2 clusters, let's plot it. This is just to verify the dataset
kmeans = KMeans(n_clusters = 2,init = 'k-means++')
y_kmeans = kmeans.fit_predict(dataset_copy)
```

```python
#Visualising kmeans
plt.scatter(dataset_copy[y_kmeans == 0,0],dataset_copy[y_kmeans == 0,1],s = 100, c = 'red',label = 'Cluster 1')
plt.scatter(dataset_copy[y_kmeans == 1,0],dataset_copy[y_kmeans == 1,1],s = 100, c = 'blue',label = 'Cluster 2')

plt.title('Clustering')
plt.xlabel('Input details')
plt.ylabel('Anamolous')
plt.legend()
plt.show()
```

```python
X = dataset.drop(['Class'],axis = 1)
y = dataset['Class']
```

```python
#Splitting into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 0,stratify = y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import StratifiedKFold,cross_val_score
from sklearn.metrics import roc_auc_score,confusion_matrix
```

```python
#Let's test the model without adding SMOTE
scoring = 'roc_auc'
models = []
models.append(('LR',LogisticRegression()))
models.append(('LDA',LinearDiscriminantAnalysis()))
models.append(('KNN',KNeighborsClassifier()))
models.append(('DT',DecisionTreeClassifier()))
models.append(('RF',RandomForestClassifier()))
models.append(('ADA',AdaBoostClassifier()))
models.append(('XGB',XGBClassifier()))
models.append(('NB',GaussianNB()))

for name, model in models:
    kfold = StratifiedKFold(n_splits = 10,random_state=0)
    cv = cross_val_score(model,X_train,y_train,scoring = scoring,n_jobs = -1)
    print('Algorithm: '+str(name))
    print('Scoring: '+str(cv.mean()))
    print('--'*30)
```

```python
#Feature Scaling 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
#LDA
lda = LinearDiscriminantAnalysis().fit(X_train_scaled,y_train)
ypred_lda = lda.predict(X_test_scaled)
print('ROC AUC Score: '+str(roc_auc_score(y_test,ypred_lda)))
```

```python
#XGBoost
xgb = XGBClassifier().fit(X_train,y_train)
ypred_xgb = xgb.predict(X_test)
print('ROC AUC Score: '+str(roc_auc_score(y_test,ypred_xgb)))
```

```python
ada = AdaBoostClassifier().fit(X_train,y_train)
ypred_ada = ada.predict(X_test)
print('ROC AUC Score: '+str(roc_auc_score(y_test,ypred_ada)))
```

All the three seems to overfit, we can tune them but let's also check the score when the minority class is over-sampled

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
#Using Logistic Regression
pipe = Pipeline([('sc',StandardScaler()),
                ('oversample',SMOTE(random_state = 0)),
                ('log',LogisticRegression(C = 1000,random_state = 0))])
sk = StratifiedKFold(n_splits=10)
params = {'log__C':[1,10,100,1000],'log__penalty':['l1','l2']}
grid = GridSearchCV(pipe,params,scoring = 'roc_auc',cv = sk)
grid.fit(X_train,y_train)
print(grid.best_score_)
print(grid.best_params_)
log = grid.fit(X_train,y_train)
ypred_log = log.predict(X_test)
cm = confusion_matrix(y_test,ypred_log)

print('ROC score: ',roc_auc_score(y_test,ypred_log))
print('Confusion Matrix:\n',cm)
```

```python
#Using Linear Discriminant Analysis
pipe = Pipeline([('sc',StandardScaler()),
                ('oversample',SMOTE(random_state = 0)),
                ('lda',LinearDiscriminantAnalysis())])
sk = StratifiedKFold(n_splits=10)
params = {'lda__n_components':[1]}
grid = GridSearchCV(pipe,params,scoring = 'roc_auc',cv = sk)
grid.fit(X_train,y_train)
lda = grid.fit(X_train,y_train)
print(grid.best_score_)
print(grid.best_params_)

ypred_lda = lda.predict(X_test)
cm = confusion_matrix(y_test,ypred_lda)

print('ROC score: ',roc_auc_score(y_test,ypred_lda))
print('Confusion Matrix:\n',cm)
```

```python
#Plotting roc curve
from sklearn.metrics import roc_curve
y_pred_prob = log.predict_proba(X_test)[:,1]
fpr, tpr, thersholds = roc_curve(y_test,y_pred_prob)
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
```

Using Logistic Regression, I got a pretty good ROC score of 0.942. This score is very good to compared to the score we achieved without oversampling
